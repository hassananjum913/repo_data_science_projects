## Time Series Plot

```python
from pylab import rcParams
rcParams['figure.figsize'] =(15,8)
df.plot()
plt.show()
```

```python
# Trend--> upward
# Seasonality --> not constant , increasing 

# So decompostion model --> multiplicative (as seasonality is not constant)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose 
```

```python
df_mult_decomp = seasonal_decompose(df, model='multplicative')
```

```python
from pylab import rcParams
rcParams['figure.figsize'] =(15,8)
df_mult_decomp.plot()
plt.show()
```

```python
# changing multiplicative model --> additive model to make seasonality constant
```

```python
df1 = df.copy()
```

```python
df1_log['#Passengers'] = np.log(df1)
```

```python
from pylab import rcParams
rcParams['figure.figsize'] =(15,8)
df1_log.plot()
plt.show()
```

```python
# side by side comaprion of multiplicative and additive model
```

```python
plt.subplot(2,1,1)
plt.title('Original Time Series plot')
plt.plot(df)

plt.subplot(2,1,2)
plt.title('Log transformed Time Series plot')
plt.plot(df1_log)

plt.tight_layout()
```
